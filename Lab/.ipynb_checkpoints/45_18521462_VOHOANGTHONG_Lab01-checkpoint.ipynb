# Sentence 12

```python
def save_file(name_file, type_file, dataframe):
    if type_file == 'json':
        dataframe.to_json(name_file + '.' + type_file, orient="split")
    elif type_file == 'csv':
        dataframe.to_csv(name_file + '.' + type_file, index = False)
    elif type_file == 'excel':
        dataframe.to_excel(name_file + '.' + type_file, sheet_name='Sheet_name_1')
    elif type_file == 'sql':
        from sqlalchemy import create_engine
        engine = create_engine('sqlite://', echo=False)
        df.to_sql('users', con=engine)
        engine.execute("SELECT * FROM users").fetchall()
    else:
        df.to_hdf(name_file + '.' + type_file, key='df', mode='w')
```

```python
save_file('dfcau12', 'csv', df)
```
