**Câu 6. Chuyển đổi “?”, hoặc “.”, hoặc “”, hoặc “_”, hoặc “Na”, hoặc “NULL”, hoặc “null”, hoặc “not”, hoặc “Not” sang NaN.**

```
import numpy as np 
df = df.replace(['??','?','.','','_','NA','NULL','null','not','Not', 'Na'], np.nan)
```

**Câu 7. Dùng phương pháp thay thế (replace) để xử lý các giá trị bị khuyết cho các
feature sau: &quot;normalized-losses&quot; (mean), &quot;stroke&quot; (choice), &quot;bore&quot; (mean),
&quot;horsepower&quot; (choice), &quot;peak-rpm&quot; (mean). Và, &quot;num-of-doors&quot; (giá trị
nhiều nhiều nhất)**

```
df['normalized-losses'] = df['normalized-losses'].replace(np.nan, df['normalized-losses'].dropna().astype(int).mean())
df['stroke'] = df['stroke'].replace(np.nan,df['stroke'].dropna().astype('float').mean())
df['bore'] = df['bore'].replace(np.nan,df['bore'].dropna().astype('float').mean())
df['horsepower'] = df['horsepower'].replace(np.nan,df['horsepower'].dropna().astype(int).mean())
df['peak-rpm'] = df['peak-rpm'].replace(np.nan,df['peak-rpm'].dropna().astype(int).mean())
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, df['num-of-doors'].value_counts().idxmax())
```

**Câu 8. Dùng phương pháp loại bỏ (drop) cho các sample đối với các feature bị khuyết.**

```
df = df.dropna()
```

```
df.dtypes
```

```
df['normalized-losses'] = df['normalized-losses'].astype('float')
df['make'] = df['make'].astype('string')
df['fuel-type'] = df['fuel-type'].astype('string')
df['engine-location'] = df['engine-location'].astype('string')
df['wheel-base'] = df['wheel-base'].astype('float')
df['bore'] = df['bore'].astype('float')
df['aspiration'] = df['aspiration'].astype('string')
df['num-of-doors'] = df['num-of-doors'].astype('string')
df['body-style'] = df['body-style'].astype('string')
df['drive-wheels'] = df['drive-wheels'].astype('string')
df['stroke'] = df['stroke'].astype('float')
df['horsepower'] = df['horsepower'].astype('float')
df['peak-rpm'] = df['peak-rpm'].astype('float')
df['price'] = df['price'].astype('int')
df['length'] = df['length'].astype('float')
df['width'] = df['width'].astype('float')
df['curb-weight'] = df['curb-weight'].astype('int')
df['engine-type'] = df['engine-type'].astype('string')
df['num-of-cylinders'] = df['num-of-cylinders'].astype('string')
df['fuel-system'] = df['fuel-system'].astype('string')
```

**Câu 11. Chuẩn hóa &quot;length&quot;, &quot;width&quot; và &quot;height&quot; trong phạm vi [0,1].**

```
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
df['length'] = min_max_scaler.fit_transform(df['length'].values.reshape(-1,1))
df['width'] = min_max_scaler.fit_transform(df['width'].values.reshape(-1,1))
df['height'] = min_max_scaler.fit_transform(df['height'].values.reshape(-1,1))
```

**Câu 12: Tạo các bin là [&#39;Low&#39;, &#39;Medium&#39;, &#39;High&#39;] cho “horsepower”**<bold text>

```
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
df['horsepower'] = pd.cut(df['horsepower'], bins, labels=['Low', 'Medium', 'High'], include_lowest=True)
```
